```
import os
import json
import pandas as pd
import sentencepiece as spm
import wget
```

Load vocab file created using [Sentence Piece](https://github.com/ionheart26/Machine-Learning-tutorial/blob/master/Sentence-Piece.ipynb)

```
vocab_file = f'{path}/kowiki.model'
vocab = spm.SentencePieceProcessor()
vocab.load(vocab_file)
```

```
filename = wget.download('https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt', f'{path}')
filename = wget.download('https://raw.githubusercontent.com/e9t/nsmc/master/ratings_test.txt', f'{path}')
for f in os.listdir(path):
  print(f)
```

```
def prepare_train(vocab, infile, outfile):
  df = pd.read_csv(infile, sep='\t', engine='python')
  with open(outfile, 'w') as f:
    for index, row in df.iterrows():
      document = row['document']
      if type(document) != str:
        continue
      instance = {'id': row['id'], 'doc': vocab.encode_as_pieces(document), 'label': row['label']}
      f.write(json.dumps(instance))
      f.write('\n')
```

```
prepare_train(vocab, f'{path}/ratings_train.txt', f'{path}/ratings_train.json')
prepare_train(vocab, f'{path}/ratings_test.txt', f'{path}/ratings_test.json')
for f in os.listdir(path):
  print(f)
```
